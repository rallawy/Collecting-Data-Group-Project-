```python
import requests 
import pandas as pd
```

```python
API_KEY = "API KEY" 
BASE_URL = 'https://api.themoviedb.org/3/movie/'
```

```python
#retrive list of 100 top rated films TMDB version
def get_top_rated():
    response_list=[]
    for i in range(1,26):
        response = requests.get(BASE_URL + 'top_rated?api_key=' + API_KEY + '&page='+str(i)).json()  
        response_list.append(response)
    return response_list
```

```python
# define column names for our new dataframe
columns = ['Top rated film', 'Numbers of males', 'Number of females', 'Non-binary','Budget','Revenue','Vote average','Vote Count','Popularity','Genre','Release Date', 'Production Country', 'Production Company']
# create dataframe with film and gender
df = pd.DataFrame(columns=columns)
response_list = get_top_rated()
for response in response_list: 
    top_rated_films = response['results']
    for film in top_rated_films:
            film_top = requests.get(BASE_URL + str(film['id']) +'/credits?api_key='+ API_KEY).json()
            film_details = requests.get(BASE_URL + str(film['id']) +'?api_key='+ API_KEY).json()
            female = 0 
            male = 0
            non_binary = 0
            not_specified = 0
            for cast in film_top['cast'][:6]:
            
                if cast['gender'] == 1:
                    female+=1 
                elif cast['gender'] == 2:
                    male+=1 
                elif cast['gender'] == 3:
                    non_binary+=1 
                else:
                    not_specified += 1
            country_data = []
            for country in film_details['production_countries']:
                country_data.append(country['name']
                )
            company_data = []
            for company in film_details['production_companies']:
                company_data.append(company['name'])
            
            df.loc[len(df)]=[film['title'],male,female,non_binary,film_details['budget'],film_details['revenue'],film_details['vote_average'],film_details['vote_count'],film_details['popularity'],film_details['genres'][0]['name'],film_details['release_date'],','.join(country_data), ','.join(company_data)]
        
```
